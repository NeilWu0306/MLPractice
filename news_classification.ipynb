```python
import glob
import os
import pandas as pd
fs=glob.glob("chinese_news_trans/*/*.txt")
cats=[]
contents=[]
for fname in fs:
    fn = os.path.split(fname)[-1]
    with open(fname,"r",encoding="utf-8")as f:
        content = f.read()
        contents.append(content)
    dn = os.path.split(fname)[0]
    dn = os.path.split(dn)[-1]
    cats.append(dn)
train_df=pd.DataFrame({
    "content":contents,
    "ans":cats
})
train_df
```

```python
fs=glob.glob("chinese_news_test/*/*.txt")
cats=[]
contents=[]
for fname in fs:
    fn = os.path.split(fname)[-1]
    with open(fname,"r",encoding="utf-8")as f:
        content = f.read()
        contents.append(content)
    dn = os.path.split(fname)[0]
    dn = os.path.split(dn)[-1]
    cats.append(dn)
test_df=pd.DataFrame({
    "content":contents,
    "ans":cats
})
test_df
```

```python
import jieba
def newscut(s):
    s=s.replace("\r","").replace("\n","")
    return " ".join(jieba.cut(s))
#apply: pandas 重要操作 拿來對整個行操作
x_train =train_df["內容"].apply(newscut)
x_test =test_df["內容"].apply(newscut)
x_test
```
