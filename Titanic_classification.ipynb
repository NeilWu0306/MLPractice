```python
import pandas as pd
train_df= pd.read_csv("titanic/train.csv",encoding="utf-8")
test_df= pd.read_csv("titanic/test.csv",encoding="utf-8")
```

```python
train_df
```

```python
test_df
```

```python
#先把答案撥出來
y_train=train_df["Survived"]
y_train
```

```python
x_train =train_df.drop(["Survived"],axis=1)
x_test =test_df
x_train
```

```python
total=pd.concat([x_train,x_test],ignore_index=True)
total
```

```python
#確認欄位缺值
total.isna().sum()
```

```python
#補數字類缺值
# PClass 理論來說應該填最常出現,因為她是類別
med = total.median()
med 
```

```python
#真的把中位數填進去
total = total.fillna(med)
total.isna().sum()
```

```python
#最常出現
most=total["Embarked"].value_counts().idxmax()
total["Embarked"]=total["Embarked"].fillna(most)
total.isna().sum()
```

```python
#稍微整理一下
total = total.drop(["Cabin","Ticket","PassengerId"],axis=1)
total
```

```python
def namecut(n):
    return n.split(",")[-1].split(".")[0].replace(" ","")
total["Name"].apply(namecut).value_counts()
```

```python
def namecut(n):
    mid = n.split(",")[-1].split(".")[0].replace(" ","")
    if mid == "Mr" or mid =="Sir":
        return "Mr"
    elif mid == "Mrs" or mid == "Mme" or mid == "Lady":
        return "Mrs"
    elif mid == "Miss" or mid == "Mlle":
        return "Miss"
    elif mid == "Master":
        return "Master"
    else:
        return None
total["Name"]=total["Name"].apply(namecut)
total
```

```python
#one hot
total=pd.get_dummies(total)
total
```

```python
x_train_process = total.iloc[:train_df.shape[0]]
x_test_process = total.iloc[train_df.shape[0]:]
x_test_process
```

```python
#https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.figure(figsize=(14,14))
temp = pd.concat([y_train,x_train_process],axis=1)
sns.heatmap(temp.corr(),
           annot=True,
           cmap="seismic")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
clf = RandomForestClassifier()
params={
    "n_estimators":range(50,70),
    "max_depth":range(5,10)
}
grid = GridSearchCV(clf,params,cv=10,n_jobs=4)
grid.fit(x_train_process,y_train)
print("最佳參數:",grid.best_params_)
print("最佳分數:",grid.best_score_)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np
clf = RandomForestClassifier(n_estimators=65, max_depth=8)
scores= cross_val_score(clf,x_train_process,y_train,cv=10,n_jobs=4)
print("10:",scores)
print("avg:",np.average(scores))
```

```python
#準備最後模型
clf = RandomForestClassifier(n_estimators=56, max_depth=9)
clf.fit(x_train_process,y_train)
pre = clf.predict(x_test_process)
pid = test_df["PassengerId"]
result=pd.DataFrame({
    "PassengerId":pid,
    "Survived":pre    
})
result.to_csv("result.csv",encoding="utf-8",index=False)
result
```

```python
pd.DataFrame({
    "Name":x_train_process.columns,
    "Imp":clf.feature_importances_
})              
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scale =scaler.fit_transform(x_train_process)
x_test_scale = scaler.transform(x_test_process)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
params={
    "n_neighbors":range(5,150)
}
grid = GridSearchCV(clf,params,cv=10,n_jobs=4)
grid.fit(x_train_scale,y_train)
print("最佳參數:",grid.best_params_)
print("最佳分數:",grid.best_score_)
```

```python
clf = KNeighborsClassifier(n_neighbors=7)
clf.fit(x_train_scale,y_train)
pre = clf.predict(x_test_scale)
pid = test_df["PassengerId"]
result=pd.DataFrame({
    "PassengerId":pid,
    "Survived":pre    
})
result.to_csv("result_knn.csv",encoding="utf-8",index=False)
result
```
